## Qualidade dos Dados
#### Custo da Sujeira: Faturamento Total antes e depois da ETL

```python
import plotly.express as px
from src.pipelines.data_quality import insight_dirty_cost
df_insights = insight_dirty_cost()
df_insights
```

Aplicação de filtro em novos df's, para gerar 2 gráficos diferentes.

```python
df_filtro_count_diff = df_insights[df_insights['Tipo'].isin(['orders_dirty', 'diff_counts'])]
df_filtro_count_type = df_insights[df_insights['Tipo'].isin(['orders_dirty', 'orders_clean'])]
df_filtro_sum = df_insights[df_insights['Tipo'].isin(['sum_dirty', 'sum_clean'])]
```

```python
map = {
    'sum_dirty': 'Faturamento Sujo',
    'sum_clean': 'Faturamento Limpo'
}

df_filtro_sum.replace(map, inplace=True)
```

```python
fig = px.bar(df_filtro_sum.sort_values(by='Valores'), 
             x='Tipo', 
             y='Valores',
             labels={'Tipo': 'Tipo de Dado (Sujo x Limpo)', 'Valores': 'Total Faturamento'},
             title='Custo da Sujeira: Faturamento Antes x Depois da Limpeza',
             color={'blue': 'Pós-Limpeza', 'red':'Pré-Limpeza'})
fig.update_layout(legend=dict(title='Tipos de Faturamento'))
fig.show()
```

Neste gráfico é possível ver que o faturamento antes da limpeza de dados era de _R$1.864.046_. Entretanto, após o tratamento, esse número passa a ser _R$1.492.185_. Representando uma diferença total de *_R$371.861_*, correspondentes à vendas perdidas por falta de informações (_como falta de identificação da loja e/ou data não registrada_). 

```python
map = {
    'orders_dirty': 'Qtd. Total de Pedidos Sujos',
    'diff_counts': 'Qtd. Perdida'
}
df_filtro_count_diff.replace(map, inplace=True)
```

```python
fig = px.pie(df_filtro_count_diff, 
             names='Tipo',
             values='Valores',
             title='Proporção Pedidos Total x Perdidos',
             hole=0.5)

fig.show()
```

Já neste gráfico PieChart, é representado o total de pedidos antes da limpeza e a porcentagem destes perdidas pós limpeza. É possível observar que cerca de *_995 pedidos_* foram perdidos por falta de informação (_sejam por erros de registro ou perda de dados_)
